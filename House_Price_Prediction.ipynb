# Data Cleaning and analysis - Task 2

```python
# filling the null values of sale price with mean
house_df['SalePrice'] = house_df['SalePrice'].fillna(house_df['SalePrice'].mean())
```

```python
# Select numeric columns
numeric_columns = house_df.select_dtypes(include=['number'])
# Count them
count = numeric_columns.shape[1]
print("Number of numerical columns:", count)
```

```python
print("Numerical Columns:")
print(numeric_columns.columns.tolist())
```

```python
print(house_df.select_dtypes(include=['number']).count())
```

```python
house_df.sort_values(by=numeric_columns.columns.tolist(),ascending=False).head(5)
```

```python
#Q6) Compare the Difference Between Year Features and SalePrice
year_features = ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt', 'YrSold']
for feature in year_features:
    if feature in house_df.columns:
        print(f"Comparison of {feature} and SalePrice:")
        print(house_df[[feature, 'SalePrice']].corr())
        print()
```

```python
Discrete_columns = house_df.select_dtypes(include=['int'])
```

```python
count = house_df.shape[1]
print("Total number of columns:", count)
```

```python
print("Discrete columns:")
print(Discrete_columns.columns.tolist())
```

```python
#Q7) Relationship Between Discrete Variables and Sales Price
for col in Discrete_columns.columns:
    if col!='SalePrice':
        print("Relationship between {col} and SalePrice:")
        print(house_df[[col, 'SalePrice']].corr())
        print()
```

```python
if 'TotalBsmtSF' in house_df.columns and 'SalePrice' in house_df.columns:
    print("Relationship between TotalBsmtSF and SalePrice:")
    print(house_df[['TotalBsmtSF', 'SalePrice']].corr())
    
```

```python
# Histogram
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.hist(house_df['SalePrice'], bins=30, color='orange', edgecolor='black')
plt.title('Distribution of Sale Price')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Logarithmic Transformation
import numpy as np
house_df['Log_SalePrice'] = np.log(house_df['SalePrice'])
```

```python
# Box Plot

import seaborn as sns
plt.figure(figsize=(10,5))
sns.boxplot(x=house_df['Log_SalePrice'], color='orange')
```

```python
#  Relationship Between Categorical Features and SalePrice
Categorical_features = house_df.select_dtypes(include=['object']).columns
for feature in Categorical_features:
    if feature in house_df.columns:
        print(f"Relationship between {feature} and SalePrice:")
        print(house_df[[feature,'SalePrice']].groupby(feature).mean())
```

```python
for col in numeric_columns:
    if col != 'SalePrice':
        print(f"Relationship between {col} and SalePrice")
        print(house_df[[col, 'SalePrice']].corr())
```

```python
continues_values = house_df.select_dtypes(include=['float']).columns
for values in continues_values:
    print(f'Comparison of {values} and SalePrice:')
    print(house_df[[values,'SalePrice']].corr())
```

```python
#filling missing or null values with mean and mode 
for col in house_df.columns:
    if house_df[col].dtype == 'object':
        m= house_df[col].mode()[0]
        house_df[col] = house_df[col].fillna(m)
    else:
        me = house_df[col].mean()
        house_df[col] = house_df[col].fillna(me)
```
